```python
# TODO:导入包
import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import seaborn as sns

# TODO:设置
os.chdir(os.getcwd())
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# TODO: 4.4.1 标准分数
# z分数/标准化值即将一组原始数据变换为均值为0方差为1的另一组数据
# 多个不同变量进行比较分析时使用，用于判断一组数据中是否有离群点
example4_1 = pd.read_csv('example4_1.csv')
z = stats.zscore(example4_1['分数'],ddof=1)
print(f"标准化分数：\n{np.round(z,4)}")
# 一般将3个标准差之外的数据确定为离群点
```

```python
# TODO: 4.4.2 极值标准化
# 将一组原始数据缩放到[0,1]的范围内，也称最小最大标准化
# 当一组数据中存在较大或则较小的极端值（离群值）时，需要进行极值标准化转换
x = example4_1['分数']
T = (x-min(x))/(max(x)-min(x))
print(f"极值标准化值：\n{np.round(T,4)}")
```

```python
# 这两种方法只是线性变换，即改变了原点和量纲，但对数据的分布性状没有改变
# 因此可以用于比较不同变量的分布特征
```

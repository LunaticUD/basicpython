```python
import numpy as np
import pandas as pd
from scipy.stats import ttest_1samp 
from statsmodels.stats.weightstats import ztest
import statsmodels.api as sm
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 小样本总体未知-t
vc = pd.read_csv('vc.csv')['vc']
t,p_value = np.round(ttest_1samp(vc,popmean=5),4)
# print(f'样本均值：{vc.mean()}；t统计量：{t}；双尾p值：{p_value}')
vc_t ={'样本均值':vc.mean(),'t统计量':t,'双尾p值':p_value}
print(pd.DataFrame(vc_t,index=['Num']))

pplot = sm.ProbPlot(vc,fit=True)
plt.subplots(1,2,figsize=(8,3))
# QQ图
ax1 = plt.subplot(121)
pplot.qqplot(line='q',ax=ax1,xlabel='期望正态值',ylabel='标准化观测值')
ax1.set_title('正态Q-Q图',fontsize=12)
# PP图
ax2 = plt.subplot(122)
pplot.ppplot(line='r',ax=ax2,xlabel='期望累积概率',ylabel='观测累积概率')
ax2.set_title('正态P-P图',fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# 大样本总体已知-z
pm250 = pd.read_csv('PM250.csv')['PM2.5值']
z,p_value = np.round(ztest(pm250,value=81,alternative='smaller'),4)
vc_t ={'样本均值':pm250.mean(),'z统计量':z,'p值':p_value}
print(pd.DataFrame(vc_t,index=['Num']))

pplot = sm.ProbPlot(pm250,fit=True)
plt.subplots(1,2,figsize=(8,3))
# QQ图
ax1 = plt.subplot(121)
pplot.qqplot(line='q',ax=ax1,xlabel='期望正态值',ylabel='标准化观测值')
ax1.set_title('正态Q-Q图',fontsize=12)
# PP图
ax2 = plt.subplot(122)
pplot.ppplot(line='r',ax=ax2,xlabel='期望累积概率',ylabel='观测累积概率')
ax2.set_title('正态P-P图',fontsize=12)
plt.tight_layout()
plt.show()
```

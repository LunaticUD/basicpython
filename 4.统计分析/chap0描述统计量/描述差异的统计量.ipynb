```python
# TODO:导入包
import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# TODO:设置
os.chdir(os.getcwd())
```

```python
# TODO: 4.2.1 极差和四分位差
example4_1 = pd.read_csv('example4_1.csv')
R = example4_1['分数'].max()-example4_1['分数'].min()
IQR = example4_1['分数'].quantile(q=0.75)-example4_1['分数'].quantile(q=0.25)
print(f'极差：{R}；\n四分位差：{IQR}；')
```

```python
# TODO: 4.2.2 方差和标准差
# 未加权
# 通过pandas计算，其ddof自由度默认为1，即n-1
var = example4_1['分数'].var()
sd = example4_1['分数'].std()
print(f'方差：{round(var,4)}；\n标准差：{round(sd,4)}；')
# 通过numpy计算，其ddof自由度默认为0，即n
var = np.var(example4_1['分数'],ddof=1)
sd = np.std(example4_1['分数'],ddof=1)
print(f'方差：{round(var,4)}；\n标准差：{round(sd,4)}；')
# 加权
# pow()用于计算数值的n次幂
example4_2 = pd.read_csv('example4_2.csv')
m = example4_2['组中值'];f = example4_2['人数']
wm = np.average(a=m,weights=f)
w_var = sum(pow(m-wm,2)*f)/(sum(f)-1)
w_sd = pow(w_var,1/2)
print(f'方差：{round(w_var,4)}；\n标准差：{round(w_sd,4)}；')
```

```python
# TODO: 4.2.3 变异系数
example4_10 = pd.read_csv('example4_10.csv')
mean = example4_10.mean()
sd = example4_10.std()
cv = sd/mean
df = pd.DataFrame({'平均数':mean,'标准差':sd,'变异系数':cv})
print(np.round(df,4))
```

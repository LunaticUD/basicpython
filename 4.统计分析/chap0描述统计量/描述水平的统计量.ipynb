```python
# TODO:导入包
import os 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# TODO:设置
os.chdir(os.getcwd())
```

```python
# 描述水平的统计量：平均数、分位数、众数
# TODO: 1.简单平均数
# 一组数据相加后除以数据个数得到的结果
expample4_1 = pd.read_csv('example4_1.csv')
print(expample4_1['分数'].mean())
print(pd.DataFrame.mean(expample4_1))
```

```python
# TODO: 2.加权平均数
# 只在所得到的数据本身是分组数据时使用，否则会有偏差
expample4_2 = pd.read_csv('example4_2.csv')
m = expample4_2['组中值']
f = expample4_2['人数']
# np.average()
# a: 要计算平均值的数组。
# axis: 指定沿哪个轴计算平均值。如果不指定，默认对整个数组进行计算。
# weights: 与数组中的值相关联的权重数组。如果未指定，所有元素的权重默认为1。
# returned: 如果为True，将返回一个包含平均值和权重和的元组；如果为False，只返回平均值。
x_bar = np.average(a=m,weights=f)
print(x_bar)
```

```python
# TODO: 3.分位数
# 将数据从小到大排序后，找到某一位置上的数值，它可以代表数据水平的高低，即分位数
# 中位数
print(expample4_1['分数'].median())
# 四分位数
print(expample4_1['分数'].quantile(q=[0.25,0.75],interpolation='linear'))
# 百分位数
print(expample4_1['分数'].quantile(q=[0.1,0.25,0.5,0.75,0.9],interpolation='linear'))
```

```python
# TODO: 4.众数
print(expample4_1['分数'].mode())
print(stats.mode(expample4_1['分数']))
```

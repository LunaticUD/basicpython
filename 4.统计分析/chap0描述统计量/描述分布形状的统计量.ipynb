```python
# TODO: 4.3.1 偏度系数和峰度系数
# 偏度是数据的不对称性，测度数据不对称性的统计量即偏度系数
# 峰度是指数据分布峰值的高低，测度数据分布峰值高低的统计量即峰度系数
example4_1 = pd.read_csv('example4_1.csv')
skew = example4_1['分数'].skew()
kurt = example4_1['分数'].kurt()
print(f'偏度系数：{round(skew,4)}；\n峰度系数：{round(kurt,4)}；')
```

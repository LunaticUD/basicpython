```python
import pandas as pd
```

```python
pinzhmidu = pd.read_csv('pinzhmidu.csv')
pinzhmidu["Treatment"] = pinzhmidu["pinzh"].astype(str) + "-" + pinzhmidu["midu"].astype(str)
print(pinzhmidu)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 构建模型（公式语法）
model1 = ols('Y ~ C(block) + C(Treatment)', data=pinzhmidu).fit()
model2 = ols('Y ~ C(block) + C(pinzh) + C(midu) + C(pinzh)*C(midu)', data=pinzhmidu).fit()
# 生成 ANOVA 表
anova_table = sm.stats.anova_lm(model2, typ=2)  # typ=2 表示使用 Type II 平方和
print(anova_table)
print(model2.summary())
```

```python
# 多重比较
import scikit_posthocs as sp

# 执行 LSD 检验（未校正 t 检验）
posthoc_density = sp.posthoc_ttest(
    pinzhmidu,
    val_col='Y',      # 数值列名
    group_col='Treatment', # 组别列名
    pool_sd=True      # 假设方差齐性
)

print("密度两两比较的 p 值矩阵:\n", posthoc_density)
```

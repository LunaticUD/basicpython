```python
# 用于不同大小数组的二进制通用函数的一组规则
import numpy as np
a = np.array([0, 1, 2])
b = np.ones((3, 3))
c = np.ones((4, 4))
print(a+a)
print(a+5)
print(a+b)
# print(a+c)   # 报错 形状不同似乎不能广播
```

```python
# 规则1：如果两个数组的维度数不同，那么小维度数组的形状将会在最左边补1.
a = np.ones((2, 3))
print(a)
b = np.arange(3)
print(b)
print(a.shape)
print(b.shape)
print(a+b)
```

```python
# 规则2：如果两个数组的形状在任何一个维度上都不匹配，那么数组的形状会沿着维度为1的维度扩展以匹配
# 另外一个数组的形状
a = np.ones((3, 1))
print(a)
b = np.arange(3)
print(b)
print(a.shape, b.shape)
# 按规则1在最左边补充1，按规则2在数组维度是1的方向扩展来匹配
print(a+b)
```

```python
# 规则3：如果两个数组的形状在任何一个维度上都不匹配并且没有一个维度上等于1，那么会引发异常.
a = np.ones((3, 2))
print(a)
b = np.arange(3)
print(b)
print(a.shape, b.shape)
# 按规则1在最左边补充1，按规则2在数组维度是1的方向扩展来匹配
# 但是维度依旧不匹配，故无法进行广播
# 但可以将b转换为列进行广播计算
print(a+b[:, np.newaxis])
```

```python
# logaddexp(a,b) ==> log(exp(a)+exp(b)) 但是比log(exp(a)+exp(b))更准确
a = np.ones((3, 1))
print(a)
b = np.arange(3)
print(b)
print(np.logaddexp(a, b))
```

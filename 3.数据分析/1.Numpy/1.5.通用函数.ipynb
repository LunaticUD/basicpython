```python
# 使用python自带的函数进行操作太缓慢，故而引入numpy的通用函数，调用C语言进行操作，反而更快
import numpy as np
x = np.arange(5)
print(x)
# 基本运算
# + 加
print(x+10)
print(np.add(x, 10))
# - 减
print(x-10)
print(np.subtract(x, 10))
# - 负数
print(-x)
print(np.negative(x))
# * 乘
print(x*10)
print(np.multiply(x ,10))
# / 除
print(x/10)
print(np.divide(x, 10))
# // 整除
print(x//2)
print(np.floor_divide(x, 2))
# ** 幂
print(x**2)
print(np.power(x, 2))
# % 取余
print(x % 3)
print(np.mod(x, 3))
```

```python
# 绝对值
# |x|
y = np.array([-1, -2, -3, 3, 5, 6])
print(np.abs(y))
```

```python
# 三角函数
z = np.linspace(0, np.pi, 9)
r = [-1, 0, 1]
print(z)
# y = sin(x)
print(np.sin(z))
# y = arcsin(x) ==> 反函数
print(np.arcsin(r))
# y = cos(x)
print(np.cos(z))
# y = arccos(x) ==> 反函数
print(np.arccos(r))
# y = tan(x)
print(np.tan(z))
# y = arctan(x) ==> 反函数
print(np.arctan(r))
```

```python
# 指数和对数
from math import log1p

x = [1, 2, 3]
print(x)
# e^x
print(np.exp(x))
# 2^x
print(np.exp2(x))
# 3^x
print(np.power(3,x))
# ln(x)
print(np.log(x))
# log2(x)
print(np.log2(x))
# log10(x)
print(np.log10(x))
# e^x + 1
print(np.expm1(x))
# ln(1 + x)
print(np.log1p(x))
```

```python
# 函数特性
# TODO y输出可以指定存储到y1中去
y1 = np.zeros(10)
y = np.arange(10)
print(np.add(y, 9, out=y1[::-1]))
```

```python
# TODO 聚合
# 一维
x = np.arange(1, 11)
print(x)
# add通用函数的reduce方法 ==> 累加
print(np.add.reduce(x))
print(np.add.accumulate(x))  # 存储每次计算的结果
# multiply通用函数的reduce方法 ==> 累积
print(np.multiply.reduce(x))
print(np.multiply.accumulate(x))    # 存储每次计算的结果
```

```python
# 二维
# 1是横轴，从左到右；0是纵轴，从上到下
x = np.arange(1, 11).reshape((5, 2))
print(x.sum())
print(x.min())
print(x.max())
print(x)
# ...
# min(),max()...类似于以下sum()
# axis=0 垂直栈
print(x.sum(axis=0))
# axis=1 水平栈
print(x.sum(axis=1))
```

```python
# TODO 外积
r = np.arange(1,10)
print(r)
# add通用函数的outer方法 ==> 广播累加
print(np.add.outer(r, r))
# multply通用函数的outer方法 ==> 广播累积
print(np.multiply.outer(r, r))
```

```python
# numpy统计函数
# python通用的是sum(),min(),max()...不做讲解
# python Numpy通用函数
# 如果数据中存在NaN,应当在函数前面加nan
# nansum(),nanmin(),...
# 求和
L = np.arange(1, 101)
print(L)
# print(np.sum(L))
# # 最小值、最大值及其索引
# print(np.min(L))
# print(np.argmin(L))
# print(np.max(L))
# print(np.argmax(L))
# 求积
print(np.prod(L))   # 与np.multiply.reduce()类似超过33后为0、超出限制、不符合规范
# 平均值 ==> (x1+x2+x3+...+xn)/n
print(np.mean(L))
# 标准差 ==> (((x1-u)^2+...+(xn-u)^2)/n)^0.5
print(np.std(L))
print(np.nanstd(L))
# 方差 ==> 标准差的平方
print(np.var(L))
# 中位数
print(np.median(L))
# 基于元素排序的统计值 ==> (P/(100/(N-1)))+1 ==> 未知？
print(np.percentile(L, 75))
```

```python
A = [0, 1, 0, 1, 0]
# 判断任意元素是否为真 ==> or关系
print(np.any(A))
# 判断所有元素是否为真 ==> and关系
print(np.all(A))
```

# np.dot()相当于计算矩阵的乘法

```python
a = np.array([[1,2],[3,4]]) 
b = np.array([[11,12],[13,14]]) 
 
# vdot 将数组展开计算内积
print (np.vdot(a,b))
```

# np.vdot()计算向量的内积

# numpy.linalg.det() 函数计算输入矩阵的行列式

```python
# numpy.linalg.solve() 函数给出了矩阵形式的线性方程的解。
# numpy.linalg.inv() 函数计算矩阵的乘法逆矩阵。
x = np.array([[1,2],[3,4]]) 
y = np.linalg.inv(x) 
print (x)
print (y)
```

```python
# 线性方程：AX=B的解
a = np.array([[1,1,1],[0,2,5],[2,5,-1]]) 
 
print ('数组 a：')
print (a)
ainv = np.linalg.inv(a) 

print ('a 的逆：')
print (ainv)
 
print ('矩阵 b：')
b = np.array([[6],[-4],[27]]) 
print (b)
 
print ('计算：A^(-1)B：')
x = np.linalg.solve(a,b) 
print (x)
```

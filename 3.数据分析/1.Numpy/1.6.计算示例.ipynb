```python
# 饲料粒度
import numpy as np

l = np.array([6.7,4.75,3.35,2.36,1.70,1.18,0.85,0.425,0.355])
m = np.array([0,0.5,15.5,32,8.5,10.5,17.5,16.5])
b = m.sum()                                                 # 质量和 ==> 下式
d = np.power(l[:0:-1]*l[7::-1],0.5)
d[0] = 0.355
d[-1] = 6.7
a = np.sum(m[::-1]*np.log10(d))                             # 上式
print("结果为：",np.power(10,a/b))
# lg^(-1)x ==> 10^x
# lgx^(-1) ==> 1/(lgx)
```

```python
# 饲料混合均匀度
data1 = np.array([90.4,89.6,90.5,89.5,96.3,87.3,87.7,92.9,92.2,92.9])
x = np.mean(data1)
s = np.std(data1,ddof=1)    # 样本标准差
print((s/x)*100)
```

```python
import numpy as np
# 生成伪随机数 ==> 与其他隔离的生成器
# 只想在某个变量用伪随机就用randomstate
i = np.random.RandomState(1)
print(i)
```

```python
# 生成固定随机数 ==> 全局的伪随机生成器
i = np.random.seed(1)
print(i)
```

```python
# 随机整数
# i = np.random.seed(1)
# np.random.RandomState(1)
print(np.random.randint(0, 10, (2, 3)))
```

```python
# 随机浮点数
print(np.random.uniform(1, 10, (2, 3)))
```

```python
# 符合正态分布样本值
print(np.random.rand(100))
```

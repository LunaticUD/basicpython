```python
import numpy as np
import pandas as pd
```

```python
# 保留索引
rng = np.random.randint(1, 6, (5, 5))
p = pd.DataFrame(rng, columns=['a', 'b', 'c', 'd', 'e'])
print(p)
print(np.exp2(p))
print(np.log(p))
```

```python
# Serise索引对齐
rng1 = pd.Series([2, 3, 4], index=['1', '2', '3'])
rng2 = pd.Series([5, 6, 7], index=['2', '3', '4'])
print(rng1, rng2)
print(rng1+rng2)  # 获得的是两个数组的并集
print(rng1.add(rng2, fill_value=0))
```

```python
# DataFrame索引对齐
ser = np.random.RandomState(42)
A = pd.DataFrame(np.random.randint(0, 20, (2, 2)),
                 columns=['A', 'B'])
B = pd.DataFrame(np.random.randint(0, 10, (3, 3)),
                 columns=['A', 'B', 'C'])
print(A, '\n', B)
fill = A.stack().mean()  # 获得的是两个数组的并集 并按相应循序相加减
print(A.add(B, fill_value=fill))
```

```python
# DataFrame与Series的运算
ser = np.random.seed(1)
A = np.random.randint(1, 10, (3, 4))
df = pd.DataFrame(A, columns=['Q', 'R', 'S', 'T'])
print(df)
print(df-df.iloc[0])
print(df.add(df["R"], axis=0))
print(df-df.iloc[1, ::2])
```

```python
r = df.iloc[1, ::2]
fill = df.mean().mean()
print(df.subtract(r,fill_value=fill))
```

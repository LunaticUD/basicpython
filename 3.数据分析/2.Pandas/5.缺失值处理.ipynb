```python
import pandas as pd
import numpy as np
```

```python
# Pandas基本上把None与NaN看成是可以等价交换的缺失值形式
# 发下缺失值
# isnull/notnull：创建一个布尔类型的掩码标签缺失值/相反操作
da = pd.Series([1, np.nan, 2, None])  # 缺失值强制转换为浮点型
df = pd.DataFrame([[1, np.nan, 2, None],
                   [np.nan, '2', 'pandas', None]],
                  columns=['A', 'B', 'C', 'D'])
print(df)
print(df.isnull())
print(df.notnull())
print(da[da.notnull()])
```

```python
# 剔除缺失值 da.dropna
# Series
print(da.dropna())
# DataFrame
print(df.dropna())
df = pd.DataFrame([[1, np.nan, 2],
                   [2, 3, 5],
                   [None, 4, 6]])
print(df.dropna(axis=0))  # 垂直栈
print(df.dropna(axis=1))  # 水平栈
print(df.dropna(axis=0, how='all'))  # how默认为any剔除带有nan整行
print(df.dropna(axis=0, thresh=3))   # thresh为非缺失值最小为多少
```

```python
# 填充缺失值da.fillna da.ffill da.bfill
# Series
da = pd.Series([1, np.nan, 2, None, 3], index=list('abcde'))
print(da)
print(da.fillna(0))
# print(da.fillna(method='ffill'))# 已废除
print(da.ffill())  # 从前往后填充
print(da.bfill())  # 从后往前填充
# DataFrame
print(df.fillna(10))  # df本身为字符串，填充的也为字符串
print(df.ffill(axis=0))
print(df.ffill(axis=1))
print(df.bfill(axis=0))
print(df.bfill(axis=1))
```

```python
import pandas as pd
```

```python
cjb = pd.read_csv('cjb.csv')
print(cjb.head())
wer = pd.Series([94,82,96,97,97,98,95,94,88])
print(wer)
```

```python
#NOTE: Series
# index 返回 Series 的索引标签（如行标签），可以是自定义或默认的整数索引
print(wer.index)
# values 返回 Series 的数值 value_counts 返回 Series 的数值次数
print(wer.values)
print(wer.value_counts())
# dtype 返回 Series 元素的数据类型
print(wer.dtype)
# name 返回或设置 Series 的名称，便于数据标识
wer.name = 'scores'
print(wer.name)
# shape 返回 Series 的维度形状，一维数组形式为 (n,)
print(wer.shape)
# size 返回 Series 中元素的总数量
print(wer.size)
#NOTE: DataFrame
# index 返回 DataFrame 的行索引标签。
print(cjb.index)
# columns 返回 DataFrame 的列名（列索引）。
print(cjb.columns)
# dtypes 返回每列的数据类型，适用于混合数据类型的表格。
print(cjb.dtypes)
# shape 返回 DataFrame 的行数和列数，格式为 (rows, columns)。
print(cjb.shape)
# values 返回 DataFrame 的数值，以二维 NumPy 数组形式存储。
print(cjb.values)
print(cjb.value_counts())
# T 返回 DataFrame 的转置（行与列互换）。
print(cjb.T)
# size 返回 DataFrame 中元素的总数量（行数 × 列数）。
print(cjb.size)
# axes 返回包含行索引和列索引的列表。
print(cjb.axes)
# ndim 返回 DataFrame 的维度数，固定为 2。
print(cjb.ndim)
```

```python
#NOTE:列选择
print(cjb[['xm','wlfk']])
#NOTE:行选择
print(cjb.loc[1:3])   # 注意 loc 包含结束索引
print(cjb.iloc[1:4])  # iloc 不包含结束索引
#NOTE:条件筛选
print(cjb[(cjb['sx'] > 90) & (cjb['yw'] < 80)])
#NOTE:字符串操作
cjb['xmlen'] = cjb['xm'].str.len()
print(cjb['xmlen'])
```

```python
df = pd.read_excel('06.xlsx',sheet_name=0)
df
```

```python
# ne()函数
df[df['行业'] != '石化']
```

```python
df[df['行业'].ne('石化')]
```

```python
# lt()函数
df[df.最高<10]
```

```python
df[df.最高.lt(10)]
```

```python
# 平均值筛选
df[df.最高<df.最高.mean()]
```

```python
df[df.最高.lt(df['最高'].mean()-23)]
```

```python
# 筛选指定行索引
df[df.index==1]
```

```python
df[df.index>2]
```

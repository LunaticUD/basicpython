```python
# 筛选字符串的数据
da = pd.read_excel('07.xlsx',sheet_name=0)
da
```

```python
df1 = da[da.loc[:,'书名'].str.len()<15]
print(df1.loc[:,'书名'])
print(df1.loc[:,'书名'].str.len())
print(da.loc[:,'书名'].str.len()<15)
```

```python
# 函数筛选
# all() axis=1行
df[(df.loc[:,['开盘','收盘']]>56).all(axis=1)]
```

```python
# any() axis=1行
df[(df.loc[:,['开盘','收盘']]>56).any(axis=1)]
```

```python
# 需注意布尔值是行还是列，这对loc取值有影响
print(da.loc[:,lambda df:df.columns.str.len()==2])
print(da.loc[lambda da:da.loc[:,'书名'].str.len()<15])
```

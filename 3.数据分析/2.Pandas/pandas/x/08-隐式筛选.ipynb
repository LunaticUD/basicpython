```python
da = pd.read_excel('07.xlsx',sheet_name=0)
da
```

```python
# contains() 筛选包含字符串的行
da[da['书名'].str.contains('编程')]
```

```python
# 运算
da[da['书名'].str.contains('编程|教程|基础')]
```

```python
da[da['书名'].str.contains('教程') & da['书名'].str.contains('基础')]
```

```python
da[da['书名'].str.contains('教程') | da['书名'].str.contains('基础')]
```

```python
# endswith() 结束字符
da[da['书名'].str.endswith('Python')]
```

```python
# startswith() 开始字符
da[da['书名'].str.startswith('Python')]
```

```python
da[da['书名'].str.startswith('Python')|da['书名'].str.endswith('Python')]
```

```python
da[da['书名'].str.startswith('Python')&da['书名'].str.endswith('Python')]
```

```python
# match()方法 多条件开始字符
da[da['书名'].str.match('数据|流畅')]
```

```python
# isnumeric() 是否全为数字 先转换.astype(str)
da[da['卖出数量'].astype(str).str.isnumeric()]
```

```python
# islower() 小写
da[da['书名'].astype(str).str.islower()]
```

```python
# isupper() 大写
da[da['书名'].astype(str).str.isupper()]
```

```python
# istitle() 检测字符串中所有的单词拼写首字母是否为大写，且其他字母为小写
da[da['书名'].astype(str).str.istitle()]
```

```python
# isin() 检查名称是否在列表中，再去筛选
df[df['名称'].isin(['中国石油','中国平安'])]
```

```python
# 取反
df[~df['名称'].isin(['中国石油','中国平安'])]
```

```python
# nlargest() 最大值操作筛选
df[df['收盘'].isin(df['收盘'].nlargest(3))]
```

```python
# nsmallest() 最小值操作筛选
df[df['收盘'].isin(df['收盘'].nsmallest(3))]
```

```python
# apply() 定义自定义函数筛选
def abc(row):
    return (row['开盘']<8) and (row['最低']<8)
df[df.apply(abc,axis=1)]
```

```python
# select_dtypes() 筛选数据类型列 object int64 float64 bool
df.select_dtypes(['object','int64','bool'])
```

```python
# filter() 筛选列明
df.filter(like='价')
```

```python
# df.filter(regex='^[^最].*')
df.filter(regex='.*量$')
```

```python
df.filter(items=['最高','最低'])
```

```python
# query() 比较运算符筛选
# df.query('最高>20')
# df.query('最高<20 and 最低<10')
df.query('8<=最高<=20')
```

```python
db = pd.read_excel('05.xlsx',sheet_name=0)
# db.query(f'英语>{db.英语.mean()}')
# db.query(f'英语>{db.英语.mean()} and 数学>{db.数学.mean()}')
yuwen = db.语文.mean()
db.query('语文 > @yuwen')
```

```python
# 链式操作
yingyu = db.英语.mean()
db.query('语文 > @yuwen').query('英语>@yingyu').query(f'数学>{db.数学.mean()}')
```

```python
# rolling() 根据样本筛选数据 window=3样本大小
dc = pd.read_excel('08.xlsx',sheet_name=0)
# dc['3日均价'] = dc['收盘'].rolling(window=3).mean()
dc['3日均价'] = dc['收盘'].rolling(window=3,center=True).mean()
dc
```

```python
# sample()根据占比筛选随机子集frac=0.3比例random_state=42种子
dc.sample(frac=0.3,random_state=42)
```

```python
# n=1 抽样数量 replace=True是否放回
dc.sample(n=1,random_state=42,replace=True)
```

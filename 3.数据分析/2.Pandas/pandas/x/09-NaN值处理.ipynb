```python
import pandas as pd
df = pd.read_excel('09.xlsx',sheet_name=0)
da = df.copy()
```

```python
# 每列的nan
nan1 = df.isna().sum()
df.loc['NAN'] = nan1
df
```

```python
# 每行的nan
nan2 = df.isna().sum(axis=1)
df['NAN'] = nan2
df
```

```python
# 每列的非nan
nan1 = df.notna().sum()
df.loc['NAN'] = nan1
df
```

```python
# 每列的非nan
nan2 = df.notna().sum(axis=1)
df['NAN'] = nan2
df
```

```python
# 每列的nan占比 len(df)总行数
nan1 = df.isna().sum()/len(df) *100
df.loc['NAN'] = nan1
df
```

```python
# 每行的nan占比 df.shape[1]总列数
nan2 = df.isna().sum(axis=1)/df.shape[1] *100
df['NAN'] = nan2
df
```

```python
# 删除NAN值 how='any' 有就删
# axis=0 行
df.dropna(axis=0,how='any')
```

```python
# axis=1 列
df.dropna(axis=1,how='any')
```

```python
# how='all' 全是nan删
# df.loc['7'] = ['NaN','NaN','NaN','NaN','NaN','NaN','NaN']
df.dropna(axis=0,how='all')
```

```python
df[['语文','数学','英语']] = df[['语文','数学','英语']].fillna(df[['语文','数学','英语']].mean().round(1))
df
```

```python
# 自定义函数填充
def fill_nan(x,col_name):
    if pd.isna(x):
        if col_name in ['姓名','性别']:
            return '不知名'
        return '缺考'
    return x
da1 = da.copy()
for col in da1.columns:
    # print(da[col])
    da1[col] = da1[col].apply(lambda x:fill_nan(x,col))

da1
```

```python
# mask填充
fill_values = {
    '1D号':'不知','姓名':'不知名','性别':'未知',
    '语文':'缺考','数学':'缺考','英语':'缺考'
}
da2 = da.copy()
for col in da2.columns:
    da2[col] = da2[col].mask(pd.isna(da2[col]),fill_values[col])

da2
```

```python
# groupby()填充
db = pd.read_excel('10.xlsx',sheet_name=0)
db
```

```python
def fill_group_nan(group):
    num_cols = ['语文','数学','英语']
    group[num_cols] = group[num_cols].fillna(round(group[num_cols].mean(),0))
    return group
dbf = db.groupby('班级').apply(fill_group_nan)
dbf
```

```python
# 分组计算
ndb = db.fillna(80)
ndb[ndb['班级'] == 101]
for col in [101,102,103]:
    dfclass = ndb[ndb['班级'] == col]
    print(dfclass)
```

```python
# 计数
# dbf[['语文','数学','英语']].count()
ndb.groupby('班级')[['语文','数学','英语']].count()
```

```python
# 求和
ndb.groupby('班级')[['语文','数学','英语']].sum()
```

```python
# 平均值
round(ndb.groupby('班级')[['语文','数学','英语']].mean(),1)
```

```python
# 班级分组计算平均值进行填充
# transform 自动对准索引，自动填写
num_cols = ['语文','数学','英语']
df_f = db.copy()
for col in num_cols:
    col_mean = round(db.groupby('班级')[col].transform('mean'),1)
    df_f[col] = db[col].fillna(col_mean)
df_f
```

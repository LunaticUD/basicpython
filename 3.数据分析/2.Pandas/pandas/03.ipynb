```python
import pandas as pd
import numpy as np
```

```python
arr = np.array([1, 2, 3])
df = pd.DataFrame({"A": arr, "B": arr.copy()}, copy=False)
df1 = pd.DataFrame({"A": arr, "B": arr.copy()})
print(df)
print(df1)
arr[0] = 0
print(df)
print(df1)
```

```python
# NOTE:创建生成
# 字典
d = {'国家': ['中国', '美国', '日本'],
     '人口': [14.33, 3.29, 1.26]}
df = pd.DataFrame(d)
df
```

```python
df = pd.DataFrame(d, index=['a', 'b', 'c'])
df
```

```python
# Series 组成的字典
d = {'one': pd.Series([1., 2., 3.], index=['a', 'b', 'c']),
     'two': pd.Series([1., 2., 3., 4.], index=['a', 'b', 'c', 'd'])}
df = pd.DataFrame(d)
df
```

```python
pd.DataFrame(d, index=['d', 'b', 'a'], columns=['two', 'three'])
```

```python
# ndarrays 或列表组成的字典
d = {'one': [1., 2., 3., 4.],
     'two': [4., 3., 2., 1.]}
print(pd.DataFrame(d))
print(pd.DataFrame(d, index=['a', 'b', 'c', 'd']))
```

```python
# 同构的数组数据
# 创建一个空的 2x3 数组
data = np.zeros((2, ), dtype=[('A', 'i4'), ('B', 'f4'), ('C', 'S10')])
print(pd.DataFrame(data))
# 给这个数据填入具体数据值
data[:] = [(1, 2., 'Hello'), (2, 3., "World")]
# 生成 DataFrame
print(pd.DataFrame(data))
```

```python
# 指定索引
pd.DataFrame(data, index=['first', 'second'])
```

```python
# 指定列名
pd.DataFrame(data, columns=['C', 'A', 'B'])
```

```python
# 字典和列表
# 定义一个字典列表
data2 = [{'a': 1, 'b': 2}, {'a': 5, 'b': 10, 'c': 20}]

# 生成 DataFrame 对象
print(pd.DataFrame(data2))
```

```python
# 指定索引
pd.DataFrame(data2, index=['first', 'second'])
```

```python
# 指定列名
pd.DataFrame(data2, columns=['a', 'b'])
```

```python
# 元组组成的字典
# 一个双索引的例子
pd.DataFrame({('a', 'b'): {('A', 'B'): 1, ('A', 'C'): 2},
              ('a', 'a'): {('A', 'C'): 3, ('A', 'B'): 4},
              ('a', 'c'): {('A', 'B'): 5, ('A', 'C'): 6},
              ('b', 'a'): {('A', 'C'): 7, ('A', 'B'): 8},
              ('b', 'b'): {('A', 'D'): 9, ('A', 'B'): 10}})
```

```python
# 由 Series 生成
s1 = pd.Series(['a', 'b', 'c', 'd', 'e'])
pd.DataFrame(s1)
```

```python
# 混杂的结构
# 从字典里生成
pd.DataFrame.from_dict(dict([('A', [1, 2, 3]), ('B', [4, 5, 6])]))
```

```python
# 如字典键为 [‘index’, ‘columns’, ‘data’, ‘index_names’, ‘column_names’]
data = {
    'index': [*'xy'], 
    'columns': [*'ab'],
    'data': [[1, 3], [2, 4]],
    'index_names': 'T',
    'column_names': 'S'
}
df = pd.DataFrame.from_dict(data, orient='tight')
df
```

```python
# 从列表、元组、ndarray 中创建
pd.DataFrame.from_records([(1, 2., b'Hello'), (2, 3., b'World')])
```

```python
#NOTE: DataFrame 的操作
d = {'one': pd.Series([1., 2., 3.], index=['a', 'b', 'c']),
     'two': pd.Series([1., 2., 3., 4.], index=['a', 'b', 'c', 'd'])}
df = pd.DataFrame(d)
df.index
```

```python
df.columns
```

```python
# 选择增加修改列
# 选择列，结果是一个 Series
df['one']
```

```python
# 定义一个固定值的列
df['foo'] = 'bar'
df
```

```python
# 定义的新列取某列的部分值
df['one_trunc'] = df['one'][:2]
df
```

```python
# 定义一个新列，由已有的两列相乘
df['three'] = df['one'] * df['two']
df
```

```python
# 新增加的列返回的是一个逻辑运算值
df['flag'] = df['one'] > 2
df
```

```python
# 在列索引位 1 处理插入名为 bar 的列，值取 df.one
df.insert(1, 'bar', df['one'])
df
```

```python
# 删除一个列
del df['two']
three = df.pop('three')
df
```

```python
# 用方法链创建新列
# 定义一个名为 rate 的新列，并给定计算公式
(df.assign(rate=df['one']/df['bar'])
    .head())
```

```python
# 可以用 lambda 进行计算，变量 x 是指整个 df 
df.assign(rate1=lambda x:x['one']/x['bar']).head()
```

```python
# 可指定多个
df.assign(rate2=lambda x:x['one']/x['bar'],
          rate3=lambda x:x['one']+x['bar']).head()
```

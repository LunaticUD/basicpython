#### 简单累计

```python
# Series -> 得到统计值
x = np.random.seed(0)
df = pd.Series(np.random.randint(1,15,6))
print(df)
print(df.mean(),df.sum())
# DataFrame 
# -> 每列统计值
df = pd.DataFrame(np.random.randint(1,15,(3,2)),index=list('abc'))
print(df)
print(df.mean(),df.sum())
# -> 每行统计值 axis默认为0 -> 垂直/列
print(df.mean(axis=1),df.sum(axis=1))
# -> 统计各项describe()
print(df.describe())
```

#### 组合操作

```python
# 读取数据
plants = pd.read_csv('./data/planets.csv')
# print(plants.dropna().describe())
print(plants)
# Groupby累计计算
df = pd.DataFrame([['a', 'man', 120, 90],
                   ['b', 'woman', 130, 100],
                   ['a', 'man', 110, 108],
                   ['a', 'woman', 120, 118]], columns=['level', 'gender', 'math','chinese'])
group = df.groupby('gender')
print(group.sum())
#TODO 1:按列取值
# 获得DataDrameGroupBy对象
print(df.groupby('gender'))            # <pandas.core.groupby.generic.DataFrameGroupBy object at 0x000001A7ACB0DDB0>
# 运用累计计算
print(df.groupby('gender').sum()) 
# 获得SeriesGroupBy对象
print(df.groupby('gender')['chinese'])   # <pandas.core.groupby.generic.SeriesGroupBy object at 0x000001A7BBE024D0>
# 运用累计计算
print(df.groupby('gender')['chinese'].sum()) 
```

#### 高效组合操作

```python
rng = np.random.RandomState(42)
df = pd.DataFrame({'key':['A','B','C','D','A','B','C','D'],
                   'data1':range(8),
                   'data2':rng.randint(20,60,8)},
                  columns=['key','data1','data2'])
print(df)
```

```python
#TODO 1:累计 字符串、函数或函数列表——>都以字符串代替/字典
print(df.groupby('key').aggregate(['min','mean','max']))
print(df.groupby('key').aggregate({'data1':'mean','data2':'sum'}))
```

```python
#TODO 2:过滤 函数不要data2方差大于10的
def ds(x):
    return x['data2'].std() > 10
print(df.groupby('key').std())
print(df.groupby('key').filter(ds))
```

```python
#TODO 3:转换 返回新的与原先形状相同的全量数据
print(df.groupby('key').transform(lambda x: x-x.mean()))
```

```python
#TODO 4:apply()方法 x['data1'].sum()是分组后的和
def nor(x):
    x['data2'] = x['data2']/x['data1'].sum()
    return x
# include_groups=False排除分组
print(df.groupby('key').apply(nor,include_groups=False))
```

#### 分割操作

```python
#TODO 1:列表、数组、Series或索引作为分组键
L = [0,1,2,0,1,0,2,1]
print(df)
print(df.groupby(L).sum())           # 列表
print(df.groupby('key').sum())       # 数组
print(df.groupby(df['key']).sum())   # 索引
```

```python
#TODO 2:字典或Series将 索引映射到分组键 'key'变为索引，再将字典映射到key索引 进行统计vowel和vow
df2 = df.set_index('key')
mapping = {'A':'vowel','B':'vowel','C':'vow','D':'vow'}
print(df2.groupby(mapping).sum())
```

```python
#TODO 3:任意Python函数 str.lower映射到key索引
print(df2.groupby(str.lower).std())
```

```python
#TODO 4:多个有效键构成的列表 之前的有效键mapping用于分组
print(df2.groupby([str.lower,mapping]).sum())
```

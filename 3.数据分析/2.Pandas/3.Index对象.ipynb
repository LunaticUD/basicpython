```python
import pandas as pd
```

```python
# Index不可变数组 不能随意修改
ind = pd.Index([2, 3, 5, 7, 11])
print(ind)
print(ind[2], ind[::2])
print(ind.size, ind.shape, ind.ndim, ind.dtype)
try:
    ind[2] = 5
except:
    print('不能更改索引')
```

```python
# Index有序集合 涉及集合操作
a = pd.Index([1, 2, 3, 4, 5])
b = pd.Index([4, 5, 6, 7, 8])
print("交集", a.intersection(b))
print("并集", a.union(b))
print("差集", a.difference(b))
```

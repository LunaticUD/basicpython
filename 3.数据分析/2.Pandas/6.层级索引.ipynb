## Â§öÁ∫ßÁ¥¢ÂºïÂàõÂª∫

```python
# PandasÂ§öÁ∫ßÁ¥¢ÂºïÔºö
# Á¨®ÂäûÊ≥ï
index = [('C', 2000), ('C', 2010),
         ('N', 2000), ('N', 2010),
         ('T', 2000), ('T', 2010)]
populations = [33871648, 37253956,
               18976457, 19378102,
               20851820, 25145561]
pop = pd.Series(populations, index=index)
print(pop)
```

```python
# Â•ΩÂäûÊ≥ï
# ÂàõÂª∫Â§öÁ∫ßÁ¥¢Âºï
index = pd.MultiIndex.from_tuples(index)
print(index)
pop = pop.reindex(index)  # Êõ¥Êñ∞Á¥¢Âºï
pop = pd.Series(populations, index=index)
print(pop)
print(pop[:, 2010])
print(pop.loc[:'N', 2000])
```

```python
# Â§öÁ∫ßÁ¥¢ÂºïËΩ¨Âåñ‰∏∫ÊôÆÈÄöÁ¥¢Âºï
pop_f = pop.unstack()
print(pop_f)
print(pop_f.stack())
```

```python
pop_df = pd.DataFrame({'total': pop, 'under18': [9267089, 9284094,
                                                 4687374, 4318033,
                                                 5906301, 6879014]})
print(pop_df)
f = pop_df['under18']/pop_df['total']
print(f)
print(f.unstack())
```

## Â§öÁ∫ßÁ¥¢ÂºïË°åÂàóËΩ¨Êç¢

```python
# Â¶ÇÊûúMultiIndex‰∏çÊòØÊúâÂ∫èÁöÑÁ¥¢ÂºïÔºåÈÇ£Â§ßÂ§öÊï∞ÂàáÁâáÊìç‰Ωú‰ºöÂ§±Ë¥•
index = pd.MultiIndex.from_product([['a', 'c', 'b'], [1, 2]])
data = pd.Series(np.random.randint(1, 10, 6), index=index)
data.index.names = ['char', 'int']
print(data)
# print(data['a':'c']) # UnsortedIndexError: 'Key length (1) was greater than MultiIndex lexsort depth (0)'Êä•Èîô
# data = data.sort_values() # ÊéíÂàóÁ≠âÁ∫ß‰∏îÊó†Â∫è
data = data.sort_index()    # ÊéíÂàóÁ¥¢Âºï
print(data)
print(data['a':'b'])
```

```python
index = pd.MultiIndex.from_product([[2014, 2015], [1, 2]],
                                   names=['year', 'visit'])
columns = pd.MultiIndex.from_product([['Bob', 'Guido', 'Sue'], ['HR', 'Temp']],
                                     names=['subject', 'type'])
data = np.random.randint(30, 50, (4, 6))
pd_da = pd.DataFrame(data, index, columns)
print(pd_da)
# stack()/unstack()ÈÄöËøálevelÂèÇÊï∞ËΩ¨Êç¢Á¥¢ÂºïÂ±ÇÁ∫ß
print(pd_da.unstack())
print(pd_da.unstack(level=0))
print(pd_da.unstack(level=1))
```

```python
data = {('C', 2000): 33871648,
        ('C', 2010): 37253956,
        ('N', 2000): 18976457,
        ('N', 2010): 19378102,
        ('T', 2000): 20851820,
        ('T', 2010): 25145561}
pop = pd.Series(data)
pop.index.names = ['state', 'year']
print(pop)
# Ë°åÂàóÊ†áÁ≠æËΩ¨Êç¢
p = pop.reset_index(name='populations')
print(p)
print(p.set_index(['state']))
print(p.set_index(['state','year']))
print(p.set_index(['state','year','populations'])) # ËøôÂ∞±Â∞ÜÊï∞ÊçÆËΩ¨Êç¢‰∏∫‰∫ÜÊ†áÁ≠æ
```

```python
index = pd.MultiIndex.from_product([[2014, 2015], [1, 2]],
                                   names=['year', 'visit'])
columns = pd.MultiIndex.from_product([['Bob', 'Guido', 'Sue'], ['HR', 'Temp']],
                                     names=['subject', 'type'])
data = np.random.randint(30, 50, (4, 6))
pd_da = pd.DataFrame(data, index, columns)
print(pd_da)
print(pd_da.mean(axis=0))
print(pd_da.mean(axis=1))
print(pd_da.sum(axis=0))
print(pd_da.sum(axis=1))
```

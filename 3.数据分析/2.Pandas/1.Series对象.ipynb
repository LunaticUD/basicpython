```python
import pandas as pd
import numpy as np
```

```python
# 检查版本
print(pd.__version__)
```

```python
# 创建Series对象————本身为一维数组
# print(pd.Series(data,index=index))
# data为列表或Numpy数组 数据必须是一维的
i = np.array([1, 2, 3, 8])
print(pd.Series(i))
print(pd.Series([5.0, 6.0, 7.0]))
```

```python
# data为标量
print(pd.Series(5, index=[1, 2, 3]))
print(pd.Series(5, index=['a', 'b', 'c']))
```

```python
# data为字典
print(pd.Series({'2': 'a', '3': 'b', '4': 'c'}))
```

```python
# 可以用显式索引指定索引筛选
print(pd.Series({'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5},
                index=['b', 'c', 'e']))
```

```python
# 属性
p = pd.Series([0.25, 0.5, 1, 1.5, 2, 2.5, 3])
print(p.values)
print(p.index)
```

```python
# 取值
print(p[3])
```

```python
# 切片
print(p[2:4])
```

```python
# 索引可以指定——是显式的
p = pd.Series([0.25,0.5,1,1.5,2],index=['a','b','c','d','e'])
print(p)
```

```python
# -------------------------------------
# 数据取值与选择
# 犹如字典
data = pd.Series([0.25,0.5,0.75,1.0],index=['a','b','c','d'])
print(data)
print(data['b'], 'a' in data)  # 取值
print(data.keys(), '\n', list(data.items()))  # 取键值
data['r'] = 55.55  # 添加
data['a'] = 99  # 修改
data['b'], data['c'] = data['c'], data['b']  # 交换
print(data)
```

```python
# 犹如一维数组
print(data['a':'b']) # 显示索引，包括最后一个索引
print(data[0:3])     # 隐式索引，不包括最后一个索引
print(data[(data > 0.3) & (data < 0.8)]) # 掩码
print(data[['a', 'd']])  # 花哨的索引
```

```python
# 区分显隐式索引，索引器
# loc属性，取值都为显性
data = pd.Series(['a','b','c'],index=[1,3,5])
print(data.loc[1])
print(data.loc[1:3])
# iloc属性，取值都为隐性
print(data.iloc[1])
print(data.iloc[1:3])
# ix属性，主要用于DataFrame，取值为混合式 == []，原则为显式大于隐式
```

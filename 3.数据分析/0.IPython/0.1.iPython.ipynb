## IPython魔法命令
```ipython
# 粘贴代码块
%paste
%cpaste
# 执行外部代码
%run
# 计算代码运行时间
%timeit
# 输出所有历史
%history
# 魔法命令帮助
%magic?
```
## 输入和输出
* In是一个列表，记录所有命令
* Out是一个字典，映射所有输出
```ipython
# 下划线获取以前输出
print(_)
print(__)
# 分号禁止输出
math.sin(2)+math.cos(2);
```

## 错误和调试
```ipython
# 控制异常
%xmode
%xmode Verbose
# 调试
%debug
# 自动调试
%pdb on
# 运行外部文件
%run -d
```
## 代码分析和计时
```ipython
# 代码段计时
%time
%timeit
# 分析整个脚本
%prun
# 逐行分析
%lprun
# 内存分析
%menit
%mprun
```

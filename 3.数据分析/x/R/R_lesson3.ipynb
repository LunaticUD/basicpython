```r
# 绘图
dose  <- c(20, 30, 40, 45, 60)
drugA <- c(16, 20, 27, 40, 60)
drugB <- c(15, 18, 25, 31, 40)
plot(dose, drugA) 
plot(dose, drugA, type="l") 
plot(dose, drugA, type="p") 
plot(dose, drugA, type="b")  
plot(dose, drugA, type="o") 
plot(dose, drugA, type="h") 
plot(dose, drugA, type="s") 
plot(dose, drugA, type="n") 
```

```r
# par()函数
## Specifying graphical parameters
# Way 1: par() function, a global setting
par()                          # default parameters
opar <- par(no.readonly=TRUE)  # makes a copy of the current settings; 
                               #no.readonly=TRUE is to extract those parameters that can be set
par(lty=2, pch=17)             # changes the default line type to dashed (lty=2) 
                               # and the default symbol for plotting points 
                               # to a solid triangle (pch=17).
plot(dose, drugA, type="b")
par(opar)                      # reset to default parameters
```

```r
# 线条
# Way 2: setting parameters directly in the plotting function
                   # but not all funtions accept all parameters 

plot(dose, drugA, type="b", lty=2, pch=24)

plot(dose, drugA, type="b", lty=3, lwd=5, 
     pch=15, cex=2) # specifying line type and symbol type, size of symbol
```

```r
# 颜色
# color
plot(dose, drugA, type="b", lty=3, lwd=3, pch=15, cex=2,col="blue")
plot(dose, drugA, type="b", lty=3, lwd=3, pch=15, cex=2,col=colors()[52])
plot(dose, drugA, type="b", lty=3, lwd=3, pch=15, cex=2,col=c("blue","red"))


n <- 30
mycolors <- rainbow(n)
pie(rep(1, n), labels=mycolors, col=mycolors)
mygrays <- gray(0:n/n)
pie(rep(1, n), labels=mygrays, col=mygrays)
```

```r
dose <- c(20, 30, 40, 45, 60)
drugA <- c(16, 20, 27, 40, 60)
drugB <- c(15, 18, 25, 31, 40)
opar <- par(no.readonly=TRUE)
par(lwd=2, cex=1.5)
par(cex.axis=.75, font.axis=3)
plot(dose, drugA, type="b", pch=19, lty=2, col="red")
plot(dose, drugB, type="b", pch=23, lty=6, col="blue", bg="green")
par(opar)
```

```r
dose <- c(20, 30, 40, 45, 60)
drugA <- c(16, 20, 27, 40, 60)
drugB <- c(15, 18, 25, 31, 40)
opar <- par(no.readonly=TRUE)
par(lwd=2, cex=1.5, font.lab=2)
plot(dose, drugA, type="b",
     pch=15, lty=1, col="red", ylim=c(0, 60),
     main="Drug A vs. Drug B",
     xlab="Drug Dosage", ylab="Drug Response")
lines(dose, drugB, type="b",
      pch=17, lty=2, col="blue")
abline(h=c(30), lwd=1.5, lty=2, col="gray")

legend("topleft", inset=.01, title="Drug Type", c("A","B"),
       lty=c(1, 2), pch=c(15, 17), col=c("red", "blue"))
par(opar)
```

```r
#########################par(mfrow)

opar <- par(no.readonly=TRUE)
par(mfrow=c(2,2))
plot(mtcars$wt,mtcars$mpg, main="Scatterplot of wt vs. mpg")
plot(mtcars$wt,mtcars$disp, main="Scatterplot of wt vs disp")
hist(mtcars$wt, main="Histogram of wt")
boxplot(mtcars$wt, main="Boxplot of wt")
par(opar)
```

```r
#########################layout() function#######
opar <- par(no.readonly=TRUE)
layout(matrix(c(1, 1, 2, 3), 2, 2, byrow = TRUE),
       widths=c(1, 1), heights=c(1, 1))  #matrix is to set the location of the figures

hist(mtcars$wt)
hist(mtcars$mpg)
hist(mtcars$disp)
par(opar)

layout(matrix(c(1, 1, 2, 3), 2, 2, byrow = TRUE),
       widths=c(1, 2), heights=c(1, 1))  #matrix is to set the location of the figures
hist(mtcars$wt)
hist(mtcars$mpg)
hist(mtcars$disp)
par(opar)
```

```r
################Creating a figure arrangement with fine control
opar <- par(no.readonly=TRUE)
par(fig=c(0, 0.8, 0, 0.8))   # fig=c(x1, x2, y1, y2)
                             # the full graph area as going  
                             # from (0,0) in the lower-left corner  
                             # to (1,1) in the upper-right corner
plot(mtcars$wt, mtcars$mpg,
     xlab = "Miles Per Gallon",
     ylab = "Car Weight")

par(fig=c(0, 0.8, 0.55, 1), new=TRUE)
boxplot(mtcars$wt, horizontal=TRUE, axes=FALSE)

par(fig=c(0.65, 1, 0, 0.8), new=TRUE)
boxplot(mtcars$mpg, axes=FALSE)

mtext("Enhanced Scatterplot", side=3, outer=TRUE, line=-2)

par(opar)
```

```r
#############ggplot2
library(ggplot2)
library(dplyr)
iris_group <- group_by(iris,Species)
head(iris_group)
iris_group_vline <- summarise(iris_group,V=mean(Petal.Length))
ggplot(data = iris, aes(x = Petal.Length)) +
  geom_histogram() +
  geom_vline(data = iris_group_vline, aes(xintercept = V, color = Species)) +  facet_grid(.~Species) +
  ggtitle("Petal Lengths of each species")


ggplot(data = iris, aes(x=Petal.Length, fill=Species)) +  geom_histogram(alpha=0.2, position="identity") +
  geom_vline(data = iris_group_vline, aes(xintercept = V, color = Species)) +  ggtitle("Petal Lengths of each species") +
  annotate("segment", x=4.8, y=14.5, xend=4.8, yend=5, size=0.5,  arrow=arrow(length=unit(.2, "cm"))) +
  annotate("text", label="Overlap", x=4.8, y=15, size=3,  fontface="bold")
```

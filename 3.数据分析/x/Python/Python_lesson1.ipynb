```python
# 3. 控制结构
x = 10
if x > 5:
    print("x is greater than 5")
elif x == 5:
    print("x equals 5")
else:
    print("x is less than 5") #条件判断

for i in range(5):  # 从 0 到 4
    print(i)

x = 0
while x < 5:  # 当 x 小于 5 时执行
    print(x)
    x += 1                        #循环
```

```python
# 8. 注释
# 这是一个单行注释
"""
这是一个
多行注释
"""

# 9. 文件操作
with open("example.txt", "r") as file:
    content = file.read()
    print(content)

# 10.文本处理
text = "Hello, World!"
# 转换为大写或小写
print(text.upper())  # 输出 "HELLO, WORLD!"
print(text.lower())  # 输出 "hello, world!"
# 去除空格
text_with_spaces = "  Hello, World!  "
print(text_with_spaces.strip())  # 输出 "Hello, World!" （去掉前后的空格）
# 分割字符串
split_text = text.split(", ")  # 以逗号加空格分割
print(split_text)  # 输出 ['Hello', 'World!']
# 字符串连接
joined_text = " ".join(split_text)  # 使用空格连接
print(joined_text)  # 输出 "Hello World!"
# 替换字符串
replaced_text = text.replace("World", "Python")
print(replaced_text)  # 输出 "Hello, Python!"
```

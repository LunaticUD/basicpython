```python
## 实例
plt.style.use('dark_background')
# 数据
df = pd.read_csv('./data/data1402.csv',header=None)
# 计数
i = {}
for j in df.values:
    i[int(j[0])] = i.get(int(j[0]),0) + 1
# 绘图
# plt.bar(i.keys(),i.values(),0.5,alpha=0.5,color='r')
## 按5分划分
score_a = []
for i in df.values:
    score_a.append(int(i[0]))
score_new = pd.Series(score_a)
bins = [60,65,70,75,80,85,90,95]
# count list
count_new = pd.cut(score_new,bins).value_counts().to_list()
x = range(len(count_new))
# xtricks list 
j=['(70, 75]', '(75, 80]', '(65, 70]', '(80, 85]', '(85, 90]', '(60, 65]', '(90, 95]']
k = [91, 66, 53, 52, 18, 18, 7]
y = range(len(k))
print(j)
print(count_new)
# 
plt.bar(j,y)
plt.show()
```

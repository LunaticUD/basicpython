```python
# 绘制函数图
x = np.linspace(-2*np.pi,2*np.pi,256)
cos,sin = 2*np.cos(2*x),np.sin(3*x)
# 基础
fig,ax=plt.subplots()
plt.plot(x,cos,'r',linewidth=1.0, linestyle="-", label='line1')
plt.plot(x,sin,'b', linewidth=1.0, linestyle="--", label='line2')
plt.legend(['2cos(2x)','sin(3x)'],loc='lower left')
# 设置边框位置
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.set_xticks([-2*np.pi,-1.5*np.pi,-1*np.pi,-0.5*np.pi,0,
               0.5*np.pi,np.pi,1.5*np.pi,2*np.pi],
              [r'$-2\pi$',r'$-\frac{3}{2}\pi$',r'$-\pi$',r'$-\frac{1}{2}\pi$',0,
               r'$\frac{1}{2}\pi$',r'$\pi$',r'$\frac{3}{2}\pi$',r'$2\pi$'])

plt.show()
```

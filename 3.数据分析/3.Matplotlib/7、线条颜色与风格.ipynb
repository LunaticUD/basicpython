```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
x = np.linspace(0,10,100)
```

### 线条颜色

```python
fig,ax = plt.subplots(figsize=(14,7))
ax.plot(x,np.sin(x-0),color='blue');
ax.plot(x,np.sin(x-1),color='g');
ax.plot(x,np.sin(x-2),color='0.75');
ax.plot(x,np.sin(x-3),color='#FFDD44');
ax.plot(x,np.sin(x-4),color=(1.0,0.2,0.3));
ax.plot(x,np.sin(x-5),color='chartreuse');
# plt.plot(x,np.sin(x-0),color='blue');
# plt.plot(x,np.sin(x-1),color='g');
# plt.plot(x,np.sin(x-2),color='0.75');
# plt.plot(x,np.sin(x-3),color='#FFDD44');
# plt.plot(x,np.sin(x-4),color=(1.0,0.2,0.3));
# plt.plot(x,np.sin(x-5),color='chartreuse');
```

### 线条的风格

```python
fig,ax = plt.subplots(figsize=(14,7))
ax.plot(x,np.sin(x-0),linestyle='-');   # 实线
ax.plot(x,np.sin(x-1),linestyle='--');  # 虚线
ax.plot(x,np.sin(x-2),linestyle='-.');  # 点划线
ax.plot(x,np.sin(x-3),linestyle=':');   # 实点线
# plt.plot(x,np.sin(x-0),linestyle='-');   # 实线
# plt.plot(x,np.sin(x-1),linestyle='--');  # 虚线
# plt.plot(x,np.sin(x-2),linestyle='-.');  # 点划线
# plt.plot(x,np.sin(x-3),linestyle=':');   # 实点线
```

### 颜色与风格组合

```python
fig,ax = plt.subplots(figsize=(14,7))
ax.plot(x,np.sin(x-0),'-b');
ax.plot(x,np.sin(x-1),'--g');
ax.plot(x,np.sin(x-2),'-.c');
ax.plot(x,np.sin(x-3),':k');
# plt.plot(x,np.sin(x-0),'-b');
# plt.plot(x,np.sin(x-1),'--g');
# plt.plot(x,np.sin(x-2),'-.c');
# plt.plot(x,np.sin(x-3),':k');
```

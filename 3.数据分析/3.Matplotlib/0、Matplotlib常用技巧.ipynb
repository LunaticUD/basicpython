```python
import numpy as np
# 导入模块
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 设置绘图样式
plt.style.use('fast')
print(plt.style.available)
```

```python
# 显示图形
# 在脚本中
# plt.show()只能写一次，一般放在最后
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
# 在ipython shell中
# 导入模块使用%matplotlib
# 每次使用plt图像都会更新一次
# plt.draw()进行强制更新
# 在ipython notebook中
# %matplotlib notebook ==> 交互式
# %matplotlib inline ==> 静态
# 运行后，每创建图形就会直接将png嵌在单元中
```

```python
# 图形保存为文件
# 使用savefig()函数
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.savefig('mysincos.png')
```

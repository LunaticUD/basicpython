```python
# 导入库
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.legend import Legend
# 设置绘图样式plt.style.use
plt.style.use('classic')
```

```python
# 离散标签 离散的图形元素
```

```python
# 绘图ax.plot接口leg = ax.legend()显示图例
x = np.linspace(0, 10, 100)
fig, ax = plt.subplots()
ax.plot(x, np.cos(x), '--g', label='Sine')
ax.plot(x, np.sin(x), '-b', label='Cosine')
# leg = ax.legend();
# 自定义标签 位置loc 外边框frameon 列数ncol
# leg = ax.legend(loc='upper right', frameon=False, ncol = 2)
# 自定义标签 圆边框fancybox 阴影shadow 透明度framealpha 文字间距borderpad
leg = ax.legend(fancybox=True, shadow=True, framealpha=0.5, borderpad=1)
```

```python
# 选择图例显示的元素
# 默认条件下，图例会忽略那些不带标签的元素
# 1、需要显示的线条传入plt.plot()
y = np.sin(x[:, np.newaxis] + np.pi + np.arange(0, 2, 0.5))
plt.plot(x, y[:, 0], label='first')
plt.plot(x, y[:, 1], label='second')
plt.plot(x, y[:, 2:])
plt.legend(framealpha=1, frameon=True);
```

```python
# 2、只为需要的线条设置标签
lines = plt.plot(x, y)
plt.legend(lines[:2], ['first', 'second']);
```

```python
# 同时显示多个图例ax.add_artist(leg)同时创建画布必不可少
fig, ax = plt.subplots(figsize=(14, 7))
lines = plt.plot(x, y)
# ax.axis('equal')
plt.legend(lines[:2], ['first', 'second'],
           loc='upper right', frameon=False)
leg = Legend(ax, lines[2:4], ['four', 'five'],
             loc='lower right', frameon=False)
ax.add_artist(leg);
```

#### 第三节、坐标上下限

```python
fig,ax = plt.subplots(figsize=(14,7))
plt.plot(x,np.sin(x));
# ax.plot(x,np.sin(x));
#plt设置坐标上下限有.xlim/.ylim或.axis([xlim,ylim])
#axes设置坐标上下限有.axis(xlim,ylim)
# # # 1
ax.set_xlim(-1,15);
ax.set_ylim(-2,2)
# plt.xlim(-1,15);
# plt.ylim(-2,2);
# # 逆序
ax.set_xlim(15,-1);
ax.set_ylim(1.2,-1.2)
# plt.xlim(15,-1);
# plt.ylim(1.2,-1.2);
# 2
ax.axis([-1,15,-2,2]);
# plt.axis([-1,15,-2,2]);
# 收紧 ==> 不留空白区域
ax.axis('tight');
# plt.axis('tight');
# 分辨率1：1 x轴单位长度与y轴单位长度相等
ax.axis('equal');
# plt.axis('equal');
```

## 第二章、散点图

```python
plt.style.use('seaborn-v0_8-whitegrid')
# 用plt.plot/ax.plot画图
x = np.linspace(0,10,30)
y = np.sin(x)
```

#### 第一节、不同属性

```python
# plt.plot(x,y,'o',color='black')
# plt.axes().plot(x,y,'o',color='black')
# # 不同的标记
for i in ['o','.',',','x','+','v','^','<','>','s','d']:
    plt.plot(np.random.randint(5),np.random.randint(5),i,label='maker={}'.format(i))
    plt.legend(numpoints=1)
    plt.xlim(-10,10)
    plt.ylim(-10,10)
```

```python
# # 组合
plt.plot(x,y,'-.pk')
# # 设置标记属性
plt.plot(x,y,'-pc',markersize=15,linewidth=4,
                    markerfacecolor='white',
                    markeredgecolor='gray',
                    markeredgewidth=2);
```

#### 第二节、用plt.scatter画图

```python
# plt.scatter(x,y,marker='o');
# 更具有灵活性
plt.scatter(np.random.randn(100),np.random.randn(100),
            c=np.random.rand(100),s=1000*np.random.rand(100),
            alpha=0.3,cmap='viridis')
plt.colorbar();
```

#### 第三节、ax.scatter画图

```python
x = np.linspace(0,20,9)
y = np.sqrt(x)
fig,ax = plt.subplots(figsize=(14,7))
CV = ['r','y','g','b','r','y','g','b','r']
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
ax.set_title('axes.scatter画图')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.scatter(x,y,s=30,c=CV,marker='x',alpha=0.5);
```

#### 第四章 多子图

```python
fig, ax = plt.subplots(2,2)
x=np.linspace(0,20,1000)
ax[0,0].plot(x,np.cos(x))
ax[0,1].plot(x,np.sin(x))
ax[1,0].plot(x,np.tan(x))
ax[1,1].plot(x,np.exp2(x))
# 刻度
for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(5))
    axi.yaxis.set_major_locator(plt.MaxNLocator(5));
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Matlab风格接口
x = np.linspace(0, 10, 100)
plt.figure()    # 创建图形
# 创建第一个图的坐标轴
plt.subplot(2, 1, 1)  # 行、列、子图编号
plt.plot(x, np.sin(x));
# 创建第二个图的坐标轴
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
# plt.gcf()   # 获取当前图形
# plt.gca()   # 获取当前坐标轴
```

```python
# 面向对象接口
# 创建图形网格
fig, ax = plt.subplots(2, figsize=(14, 7))
ax[0].plot(x, np.sin(x));
ax[1].plot(x, np.cos(x));
```

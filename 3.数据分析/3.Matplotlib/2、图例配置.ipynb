```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 支持中文
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
x = np.linspace(0, 10, 1000)
```

### 简单图例

```python
fig, ax = plt.subplots(figsize=(14,7))
ax.plot(x, np.sin(x), '-b', label='sin(x)')
ax.plot(x, np.cos(x), '--r', label='cos(x)')
ax.axis('equal')
ax.legend();
```

```python
# 图例位置、外边框、标签列数、圆边框、阴影、外边框透明度、文字间距
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(x, np.sin(x), '-b', label='sin函数')
ax.plot(x, np.cos(x), '--r', label='cos函数')
ax.axis('equal')
# ax.legend(loc='upper left',frameon=False);
# ax.legend(loc='lower center',frameon=False,ncol=2);
ax.legend(loc='upper right', frameon=True, ncol=2, fancybox=True,
          shadow=True, framealpha=1, borderpad=3);
```

```python
print(help(ax.legend()))
```

### 图例显示信息

```python
fig, ax = plt.subplots(figsize=(14, 7))
# 需要显示的线条传入plt.legend
y = np.exp(x[:, np.newaxis]+np.linspace(0, 10, 10))
line = ax.plot(x, y)
plt.legend(line[:2], ['a', 'b']);
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
# 需要显示图例中的设置标签 图例会忽略不带标签的元素
ax.plot(x, np.sin(x-0), color='blue', label='Blue')
ax.plot(x, np.sin(x-3), color='#FFDD44', label='Yellow')
ax.plot(x, np.sin(x-1), color='g')
ax.plot(x, np.sin(x-2), color='0.75')
ax.legend(loc='upper right', framealpha=1,
          frameon=True, shadow=True, borderpad=3);
```

### 显示不同尺寸的点

```python
cities = pd.read_csv('./data/california_cities.csv')
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
# c=np.log10(population)使用对数尺度
# linewidth=0：设置散点的边缘线宽度为0，意味着散点之间没有边框线
ax.scatter(lat, lon, label=None, c=np.log10(population),
           cmap='viridis', s=area, linewidths=0, alpha=0.5)
ax.set_xlabel('longitude')
ax.set_ylabel('latiyude')
ax.set_title('California Cities')
# 设置图例
for area in [100, 300, 500]:
    ax.scatter([], [], c='k', alpha=0.3, s=area, label=str(area)+'km$^2$')
# scatterpoints=1只显示一个点 labelspacing=1 表示标签之间的垂直间距
ax.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Area');
```

### 显示多个图例

```python
from matplotlib.legend import Legend
# 显示多个图例
fig, ax = plt.subplots(figsize=(14, 7))
y = np.exp(x[:, np.newaxis]+np.linspace(0, 10, 10))
line = ax.plot(x, y)
ax.legend(line[:2], ['a', 'b'], loc='upper right');
leg = Legend(ax, line[2:4], ['c', 'd'], loc='upper left')
ax.add_artist(leg);
```

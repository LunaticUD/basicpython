```python
# Basic
import matplotlib.pyplot as plt
import numpy as np
# Data
x = np.random.normal(size=50000)
y = x * 3 + np.random.normal(size=50000)
# A histogram 2D
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)
# Add a basic title
plt.title("A 2D histogram")
# Show the graph
plt.show()
```

```python
# Position
#  ‘center’, ‘left’, and ‘right’
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)
plt.title("A 2D histogram", loc='left')
plt.show()
# horizontalalignment ('center', 'right', 'left') 
# verticalalignment ('top', 'bottom', 'center', 'baseline')
fig, ax = plt.subplots()
ax.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)
ax.set_title("A 2D histogram", loc='left',horizontalalignment='center', verticalalignment='center')
plt.show()
```

```python
# Font
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)
# Italic and purple, size 20
plt.title("A 2D histogram", fontsize=20, fontweight=1000, color='purple', loc='left', style='italic' )
# to write in bold -->latex
plt.title( "$\mathbf{write in bold}$" , loc="right")
plt.show()
```

```python
# Multi-lines \n
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)
plt.title('A 2D histogram\nwith the Reds palette', loc='left')
# Use plt.subplots_adjust(top=0.7) if you need to save your image. It avoids title to be truncated.
plt.show()
```

```python
# Multi-title
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)
# Fist title
plt.title("A 2D histogram", loc='left', fontsize=18)
# Second title
plt.title("made in Python", loc='right', fontsize=13, color='grey', style='italic')
plt.show()
```

```python
# Math latex
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)
plt.title("$\mathcal{A}\mathrm{sin}(2 \omega t)$", fontsize=22)
plt.show()
```

```python
# Suptitle 超标题
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)
# Add suptitle above the title
plt.suptitle("A 2D histogram\n", fontsize=18, y=1.02)
# Add title
plt.title("Realized by the Python Graph Gallery", color="grey", style='italic')
plt.show()
```

```python
# Margin 边缘 y
# 标题和图片的距离 pad
plt.hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds)
# Add a title
plt.title("A 2D histogram", y=1.05, pad=25)
plt.show()
```

# Annotation

```python
# Text
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.DataFrame({'x_pos': range(1,101), 'y_pos': np.random.randn(100)*15+range(1,101) })
plt.plot('x_pos', 'y_pos', data=df,  linestyle='none', marker='o')

# Annotate with text + Arrow
plt.annotate(
    # Label and coordinate
    'This point is interesting!', 
    # 数据位置
    xy=(25, 50), 
    # 文本位置
    xytext=(0, 80),
    # 箭头
    arrowprops=dict(facecolor='black', shrink=0.05)
)
plt.show()
```

```python
# Math
fig,ax = plt.subplots()
plt.plot( 'x_pos', 'y_pos', data=df, linestyle='none', marker='o')
# Annotation
ax.text(20, 100, r'equation: $\sum_{i=0}^\infty x_i$', fontsize=20)
plt.show()
```

```python
# Rectangle
import matplotlib.patches as patches
fig,ax = plt.subplots()
ax.plot( 'x_pos', 'y_pos', data=df, linestyle='none', marker='o')
 
# Add rectangle
ax.add_patch(
    patches.Rectangle(
        (20, 25), # (x,y)
        width=50, # width
        height=50, # height
        # You can add rotation as well with 'angle'
        alpha=0.3, facecolor="red", edgecolor="black", linewidth=3, linestyle='solid'
    )
)
plt.show()
```

```python
# Circle
# Plot
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.plot( 'x_pos', 'y_pos', data=df, linestyle='none', marker='o')
 
# Annotation
ax1.add_patch(
    patches.Circle(
        (40, 35),           # (x,y)
        radius=20,                    # radius
        alpha=0.3, facecolor="green", edgecolor="black", linewidth=1, linestyle='solid'
    )
)
plt.show()
```

```python
# Ellipse
# Plot
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.plot( 'x_pos', 'y_pos', data=df, linestyle='none', marker='o')

# Annotation
ax1.add_patch(
    patches.Ellipse(
        (40, 35), # (x,y)
        width=30, # width
        height=100, # height
        angle=-45, # radius
        alpha=0.3, facecolor="green", edgecolor="black", linewidth=1, linestyle='solid'
    )
)
plt.show()
```

```python
# Segment
# Basic chart
plt.plot( 'x_pos', 'y_pos', data=df, linestyle='none', marker='o')

# Annotation
plt.plot([80, 40], [30, 90], color="skyblue", lw=5, linestyle='solid')
plt.show()
```

```python
# Vertical and Horizontal Lines
fig,ax = plt.subplots()
plt.plot( 'x_pos', 'y_pos', data=df, linestyle='none', marker='o')
# Annotation
ax.axvline(40, color='r')
ax.axhline(40, color='green')
# Show the graph
plt.show()
```

# Subplots

```python
# 2 Columns
# 3-digit integer
# libraries and data
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
df=pd.DataFrame({'x_values': range(1,101), 'y_values': np.random.randn(100)*15+range(1,101), 'z_values': (np.random.randn(100)*15+range(1,101))*2 })
 
# Cut your window in 1 row and 2 columns, and start a plot in the first part
plt.subplot(121)
plt.plot( 'x_values', 'y_values', data=df, marker='o', alpha=0.4)
plt.title("A subplot with 2 lines")
 
# And now add something in the second part:
plt.subplot(122)
plt.plot( 'x_values','z_values', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)
 
# Show the graph
plt.show()
```

```python
# 2 Rows
# fist line:
plt.subplot(211)
plt.plot( 'x_values', 'y_values', data=df, marker='o', alpha=0.4)
plt.title("A subplot with 2 lines")
 
# second line
plt.subplot(212)
plt.plot( 'x_values','z_values', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)

# Show the graph
plt.show()
```

```python
# Shared Axis
# initialise the figure. here we share X and Y axis
# sharex=True, sharey=True 共享xy
fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True, sharey=True)
axes[0].plot( 'x_values', 'y_values', data=df, marker='o', alpha=0.4)
axes[1].plot( 'x_values','z_values', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)
axes[0].title.set_text('These 2 plots have the same limit for the Y axis')
plt.show()
```

```python
# 2 Rows and 2 Columns
plt.subplot(221)
plt.plot( 'x_values', 'y_values', data=df, marker='o', alpha=0.4)
plt.subplot(222)
plt.plot( 'x_values','z_values', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)
plt.subplot(223)
plt.plot( 'x_values','z_values', data=df, linestyle='none', marker='D', color="green", alpha=0.3)
plt.subplot(224)
plt.plot( 'x_values','z_values', data=df, marker='o', color="grey", alpha=0.3)
plt.show()
```

```python
# Title
# initialize a figure
fig=plt.figure()
plt.subplot(221)
plt.plot( 'x_values', 'y_values', data=df, marker='o', alpha=0.4)
plt.title('title of fig A', fontsize=12, color='grey', loc='left', style='italic')
plt.subplot(222)
plt.plot( 'x_values','z_values', data=df, linestyle='none', marker='o', color="orange", alpha=0.3)
plt.title('title of fig A', fontsize=12, color='grey', loc='left', style='italic')
plt.subplot(223)
plt.plot( 'x_values','z_values', data=df, linestyle='none', marker='D', color="green", alpha=0.3)
plt.title('title of fig A', fontsize=12, color='grey', loc='left', style='italic')
plt.subplot(224)
plt.plot( 'x_values','z_values', data=df, marker='o', color="grey", alpha=0.3)
plt.title('title of fig A', fontsize=12, color='grey', loc='left', style='italic')
# Add a title:
plt.suptitle('A title common to my 4 plots', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Custom Proportion
# subplot2grid () 函数
# shape:其中第一个数字是行数，第二个数字是列数。
# location (loc):它用于指定放置子图的行号和列号。还需要注意的是，索引从0开始。(0,0)是网格的第一行第一列的单元格。
# rowspan:就可以使用这个参数展开选择以占用更多的行。
# colspan:类似于rowspan，用于扩展选择以占用更多的列。
# Plot 1
ax1 = plt.subplot2grid((2, 2), (0, 0), rowspan=2)
ax1.plot( 'x_values', 'y_values', data=df, marker='o', alpha=0.4)
# Plot 2
ax2 = plt.subplot2grid((2, 2), (0, 1), rowspan=1)
ax2.plot( 'x_values','z_values', data=df, marker='o', color="grey", alpha=0.3)
# Plot 3
ax3 = plt.subplot2grid((2, 2), (1, 1), rowspan=1)
ax3.plot( 'x_values','z_values', data=df, marker='o', color="orange", alpha=0.3)
plt.show()
```

```python
# Remove
# ax.remove
import palmerpenguins
import matplotlib.pyplot as plt
penguins = palmerpenguins.load_penguins().dropna()
penguins.head()
```

```python
COLUMNS = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
COLORS = ["#386cb0", "#fdb462", "#7fc97f"]
SPECIES = ["Adelie", "Gentoo", "Chinstrap"]
# A layout of 4x4 subplots
fig, axes = plt.subplots(4, 4, figsize = (12, 8), sharex="col", tight_layout=True)

for i in range(len(COLUMNS)):
    for j in range(len(COLUMNS)):
        # If this is the lower-triangule, add a scatterlpot for each group.
        if i > j:
            for species, color in zip(SPECIES, COLORS):
                data = penguins[penguins["species"] == species]
                axes[i, j].scatter(COLUMNS[j], COLUMNS[i], color=color, alpha=0.5, data=data)
                
        # If this is the main diagonal, add histograms
        if i == j:
            for species, color in zip(SPECIES, COLORS):
                data = penguins[penguins["species"] == species]
                axes[i, j].hist(COLUMNS[j], bins=15, alpha=0.5, data=data)
```

```python
# Remove unused axes
for i in range(len(COLUMNS)):
    for j in range(len(COLUMNS)):
        # If on the upper triangle
        if i < j:
            axes[i, j].remove()

# See the chart now
fig
```

# Customizing the legend

```python
import palmerpenguins
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = palmerpenguins.load_penguins()
```

```python
FLIPPER_LENGTH = data["flipper_length_mm"].values
BILL_LENGTH = data["bill_length_mm"].values
SPECIES = data["species"].values
SPECIES_ = np.unique(SPECIES)
COLORS = ["#1B9E77", "#D95F02", "#7570B3"]
fig, ax = plt.subplots(figsize=(8,8))
for species, color in zip(SPECIES_, COLORS):
    idxs = np.where(SPECIES == species)
    # No legend will be generated if we don't pass label=species
    ax.scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, alpha=0.7
    )
    
ax.legend()
plt.show()
```

```python
# The markers we use in the scatterplot
MARKERS = ["o", "^", "s"] # circle, triangle, square
fig, ax = plt.subplots(figsize=(8,8))
for species, color, marker in zip(SPECIES_, COLORS, MARKERS):
    idxs = np.where(SPECIES == species)
    ax.scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, marker=marker, alpha=0.7
    )
    
ax.legend()
plt.show()
```

```python
# The legend position
MARKERS = ["o", "^", "s"] # circle, triangle, square
fig, ax = plt.subplots(figsize=(8,8))
for species, color, marker in zip(SPECIES_, COLORS, MARKERS):
    idxs = np.where(SPECIES == species)
    ax.scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, marker=marker, alpha=0.7
    )
ax.legend(loc="lower right") 
plt.show()
```

```python
# Legend outside the plot
fig, ax = plt.subplots(figsize=(8,8))
for species, color in zip(SPECIES_, COLORS):
    idxs = np.where(SPECIES == species)
    ax.scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, alpha=0.7
    )
fig.subplots_adjust(right=0.8)
# Add the legend
# Pass `fig.transFigure` as the bounding box transformation 'bbox_transform'
# loc="center left" and bbox_to_anchor=(0.8, 0.5) indicate the left border
# of the legend is placed at 0.8 in the x axis of the figure, and it is
# vertically centered at y = 0.5
ax.legend(
    loc="center left",
    # bbox_to_anchor=(0.8, 0.85), # 锚点
    bbox_to_anchor=(1, 0.95),
    borderaxespad=0
    # bbox_transform=fig.transFigure
    # fig.transFigure相对整个图形，适合全局定位。
    # ax.transAxes相对某个子图（Axes），适合局部定位。
    # ax.transData直接与绘图数据相关联
)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
for species, color in zip(SPECIES_, COLORS):
    idxs = np.where(SPECIES == species)
    ax.scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, alpha=0.7
    )
# Make room on top now
fig.subplots_adjust(top=0.8)
ax.legend(
    loc="lower center", # "upper center" 线之下
    ncol=3,
    bbox_to_anchor=(0.5, 0.8),
    bbox_transform=fig.transFigure
)
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(8, 12), tight_layout=True)
for species, color in zip(SPECIES_, COLORS):
    idxs = np.where(SPECIES == species)
    axes[0].scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, alpha=0.7
    )
    
    axes[1].scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, alpha=0.7
    )

# borderaxespad = 0 消除图例与坐标轴之间的填充空间
axes[0].legend(
    loc="center left", 
    ncol=3,
    bbox_to_anchor=[0, 1.05],
    borderaxespad=0,
)

axes[1].legend(
    loc="center right", 
    ncol=3,
    bbox_to_anchor=[1, 1.05],
    borderaxespad=0,
);
```

```python
# Customize labels
# legend.get_texts()
fig, ax = plt.subplots(figsize=(8, 6))
for species, color in zip(SPECIES_, COLORS):
    idxs = np.where(SPECIES == species)
    ax.scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, alpha=0.7
    )
legend = ax.legend(loc="lower right")
# Iterate over texts.
# Method names are quite self-describing
for text in legend.get_texts():
    text.set_color("#b13f64")
    text.set_fontstyle("italic")
    text.set_fontweight("bold")
    text.set_fontsize(14)
    text.set_rotation(20) # 旋转
    text.set_backgroundcolor("#d1afe8")
    text.set_alpha(0.3)
plt.show()
```

```python
# Customize the title
fig, ax = plt.subplots(figsize=(8, 6))
for species, color in zip(SPECIES_, COLORS):
    idxs = np.where(SPECIES == species)
    ax.scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, alpha=0.7
    )

# Add a title
legend = ax.legend(
    loc="lower right", 
    title_fontsize=13, 
    frameon=False # 无框
)
# Set title manually. Just specify the text.
legend.set_title("Species")
# Object of class 'matplotlib.text.Text'. 
# We can use any of the methods we have used for the label texts above.
title = legend.get_title()
title.set_color("#ca699d")
title.set_weight("bold")
title.set_size(15)
title.set_x(-60) # 设置位置
title.set_y(-60)
title.set_rotation(90) # 设置旋转
plt.show()
```

```python
# Customize the handles
fig, ax = plt.subplots(figsize=(8,8))
for species, color in zip(SPECIES_, COLORS):
    idxs = np.where(SPECIES == species)
    ax.scatter(
        FLIPPER_LENGTH[idxs], BILL_LENGTH[idxs], label=species,
        s=50, color=color, alpha=0.7
    )

legend = ax.legend(loc="lower right", fontsize=14, markerscale=4, labelspacing=2, frameon=False)
handles = legend.legend_handles

# There are many more hatches available in matplotlib
hatches = ["+", "x", "o"]
colors = ["#f6d2a9", "#ea8171", "#b13f64"]

for i, handle in enumerate(handles):
    handle.set_edgecolor("#6c2167") # 设置边缘
    handle.set_facecolor(colors[i]) # 设置颜色
    handle.set_hatch(hatches[i]) # 设置填充
    handle.set_alpha(0.7)

plt.show()
```

```python
# borderaxespad: The pad between the axes and legend border, in font-size units.
# borderpad: The fractional whitespace inside the legend border, in font-size units.
# columnspacing: The spacing between columns, in font-size units.
# handlelength: The length of the legend handles, in font-size units.
# handletextpad: The pad between the legend handle and text, in font-size units.
# labelspacing: The vertical space between the legend entries, in font-size units.
# markerscale: The relative size of legend markers compared with the originally drawn ones.
```

# ADD hatchs and patterns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.bar(
   height=[50, 70, 30],
   x=['Group A', 'Group B', 'Group C'],
   hatch=['*', 'O', '+'],
   color=['darkred', 'lightgreen', 'skyblue']
)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(
   x, y,
   color='darkred',
)
plt.fill_between(
   x, y,
   color='red',
   hatch='///',
   alpha=0.4,
   edgecolor='black'
)
plt.show()
```

```python
url = 'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/iris.csv'
df = pd.read_csv(url)
grouped = df.groupby('species')['sepal_length']

# Init a figure and axes
fig, ax = plt.subplots(figsize=(8, 8))

# Create the plot with different colors for each group
boxplot = ax.boxplot(
    x=[group.values for name, group in grouped],
    patch_artist=True,
)

# Define colors for each group
colors = ['orange', 'purple', 'green']
hatchs = ['//', '//', '//']

# Assign colors to each box in the boxplot
for box, color, hatch in zip(boxplot['boxes'], colors, hatchs):
    box.set_facecolor(color)
    box.set(hatch=hatch)

# Display it
plt.show()
```

```python
# library & dataset
import seaborn as sns
df = sns.load_dataset('iris')
# Figure size
fig,ax = plt.subplots()
sns.set_theme(rc={'figure.figsize':(10,10)})
sns.set_theme(style="darkgrid")
sns.kdeplot(df['sepal_width'],ax=ax)
ax.set_title('A distribution chart made with Seaborn')
plt.show()
```

```python
# Marginal Plot
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
 
# Custom the inside plot: options are: “scatter” | “reg” | “resid” | “kde” | “hex”
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='scatter')
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='hex')
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='kde')

plt.show()
```

```python
# Violin plot
# calculate medians and number of observations
medians = df.groupby(['species'])['sepal_length'].median().values
nobs = df['species'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n: " + i for i in nobs]
  
sns.set_theme(style="white")
ax = sns.violinplot(x="species", y="sepal_length", data=df)
 
# Add text to the figure
pos = range(len(nobs))
for tick, label in zip(pos, ax.get_xticklabels()):
   ax.text(pos[tick], medians[tick] + 2, nobs[tick],
            horizontalalignment='center',
            size='small',
            color='black')
plt.show()
```

```python
# heatmap
# libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pypalettes import load_cmap
cmap = load_cmap('Acanthurus_leucosternon', cmap_type='continuous')
df = pd.DataFrame(
    np.random.random((10,10)),
    columns=["a","b","c","d","e","f","g","h","i","j"]
)
```

```python
# plot a heatmap with annotation
sns.heatmap(df, annot=True, 
            annot_kws={"size": 7},
            cmap=cmap,
            linewidths=2, 
            linecolor='black',
            yticklabels=False,
            cbar=True
            )
plt.show()
```
